```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv("MODULO7_PROJETOFINAL_BASE_SUPERMERCADO - MODULO7_PROJETOFINAL_BASE_SUPERMERCADO (1).csv.csv")

df.head(10)
```

```python
# Trazendo as médias e as medianas dos preços por categoria
# MÉDIA
media_preco_por_categoria = df.groupby("Categoria")["Preco_Normal"].mean().reset_index()
fig = px.bar(media_preco_por_categoria, x="Preco_Normal", y="Categoria", orientation="h",
             title="Média de preços para cada categoria",
             labels={"Preco_Normal": "Preço", "Categoria": "Categoria"},
             color="Preco_Normal",
             width=800)
fig.show()
```

```python
# Trazendo as médias e as medianas dos preços por categoria
# MEDIANA
media_preco_por_categoria = df.groupby("Categoria")["Preco_Normal"].median().reset_index()
fig = px.bar(media_preco_por_categoria, x="Preco_Normal", y="Categoria", orientation="h",
             title="A Mediana de preços para cada categoria",
             labels={"Preco_Normal": "Preço", "Categoria": "Categoria"},
             color="Preco_Normal",
             width=800)
fig.show()
```

```python
# Trazendo os devios padrão dos preços por categoria
desvio_preco_por_categoria = df.groupby("Categoria")["Preco_Normal"].std().reset_index()
fig = px.bar(desvio_preco_por_categoria, x="Preco_Normal", y="Categoria", orientation="h",
             title="desvio padrão dos preços para cada categoria",
             labels={"Preco_Normal": "Preço", "Categoria": "Categoria"},
             color="Preco_Normal",
             width=800)
fig.show()
```

```python
# Boxplot da distribuição do Preco_Normal para a categoria que tem o maior desvio padrão ( lacteos )
categoria_mais_desvio_padrao = df.loc[df["Categoria"] == "lacteos"]
fig = px.box(categoria_mais_desvio_padrao, y="Preco_Normal", title="Boxplot da Categoria: Lácteos")

fig.show()
```

```python
# Gráfico de barras onde temos a média de descontos por categoria
media_desconto_por_categoria =  df.groupby("Categoria")["Preco_Desconto"].mean().reset_index()
fig = px.bar(media_desconto_por_categoria, y="Categoria", x="Preco_Desconto", orientation="h",
             title="Média de Desconto por categoria",
             labels={"Categoria": "Categoria", "Preco_Desconto": "Preco_Desconto"},
             color="Preco_Desconto",
             width=800)
fig.show()
```

```python
# Gráfico de mapa interativo agrupando os dados por categoria, marca e trazendo a média de desconto.
desconto_de_categoria_marca = df.groupby(["Categoria", "Marca"])["Preco_Desconto"].mean().reset_index()

fig = px.treemap(desconto_de_categoria_marca,
                 path=["Categoria", "Marca"],
                 values="Preco_Desconto",
                 title="Média de desconto por Marca e Categoria",
                 color="Marca",
                 width=1100)
fig.show()
```
